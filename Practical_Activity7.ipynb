```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plot 
```

```python
football=pd.read_csv("C:\\Users\\user\\Downloads\\dataset - 2020-09-24.csv")
```

Display columns

```python
print(football.columns)
```

```python
sns.set_theme()
```

Apply the style 

```python
sns.set_style("darkgrid")
```

plot the graph now

```python

sns.lineplot(x='Jersey Number',y='Age',data=football)
plot.show()
```

Second line plot

```python

sns.lineplot(x='Club',y='Nationality',data=football)
plot.show()
```

Distribution plot for line graph number 1

```python
sns.displot(football['Age'])
```

Distribution plot for line graph number 2

```python
sns.displot(football['Wins'])
```

KDE plot for graph number 1

```python
sns.kdeplot(football['Age'],fill=True)
```

This is the kde graph for line plot number 1 but using second column

```python
sns.kdeplot(football['Jersey Number'],fill=True)
plot.show()
```

Rug Plot graph for line plot number 2 but using column using Club

```python
sns.rugplot(football['Club'])
plot.show()
```

Same plot using different Column on line plot number 1

```python
sns.rugplot(football['Nationality'])
plot.show()
```

Box plot for line plot number 1

```python
sns.boxplot(x='Jersey Number',y='Age',data=football)
plot.show()
```

Box plot for line plot number 2

```python
sns.boxplot(x='Club',y='Nationality',data=football)
plot.show()
```

Violin plot graph for line plot number 1

```python
sns.violinplot(x='Jersey Number',y='Age',data=football)
plot.show()
```

Violin plot for line graph number 2

```python
sns.violinplot(x='Club',y='Nationality',data=football)
plot.show()
```

Swarmplot for line plot number 1

```python
sns.swarmplot(x="Jersey Number",y="Age",data=football)
plot.show()
```

Swarm plot for line plot number 2

```python
sns.swarmplot(x="Club",y="Nationality",data=football)
plot.show()
```

Pair plot for line plot graph number 1

```python
sns.pairplot(football, hue='Nationality')
plot.show()
```

Jointplot for line plot number 1

```python
sns.jointplot(x="Jersey Number",y="Age",data=football,kind="scatter")
plot.show()
```
